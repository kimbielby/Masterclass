```python
import numpy as np
import matplotlib.pyplot as plt
import colorsys
```

```python
hue = 45
h = hue / 360   # convert to 0-1 range

# Define sat and lum levels
sat_levels = [0.3, 0.5, 0.7]
lum_levels = [0.3, 0.5, 0.7]
```

```python
# Build colour grid
colours = []
for l in lum_levels:
    row = []
    for s in sat_levels:
        rgb = colorsys.hls_to_rgb(h, l, s)
        row.append(rgb)
    colours.append(row)

```

```python
# Display with matplotlib
fig, ax = plt.subplots(figsize=(6, 3))
ax.imshow(colours, extent=[0, 3, 0, 3])
ax.set_xticks(np.arange(0.5, 3.5, 1))
ax.set_xticklabels([f"S={s}" for s in sat_levels])
ax.set_yticks(np.arange(0.5, 3.5, 1))
ax.set_yticklabels([f"L={l}" for l in reversed(lum_levels)])
ax.set_title(f"Hue {hue} in HSL - Variation of Saturation and Luminosity")
plt.show()

```
